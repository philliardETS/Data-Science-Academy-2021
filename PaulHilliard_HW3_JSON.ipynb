```python
import numpy as np
import pandas as pd
import json
```

```python
with open('week3_homework.json') as f:
    file=json.load(f)
file.values()
```

```python
#grab the eventSequence
dfeventSequence=pd.DataFrame(file['gameLog'][0]['eventSequence'])
dfeventSequence
```

```python
#this gives us the keys
file['gameLog'][0].keys()
```

```python
#possible merge
dfgameLog = pd.DataFrame(file['gameLog'])
dfplayerID = pd.DataFrame(file['gameLog'][0]['playerID'])
dfeventSequence = pd.DataFrame(file['gameLog'][0].get('eventSequence'))
x = dfgameLog.merge(dfplayerID, how="cross").merge(dfeventSequence, how="cross")
x
```

```python
playerID0=pd.DataFrame(file['gameLog'][0]['playerID'])
playerID1=playerID0.tail(1)
playerID0=playerID0.head(1)
playerID0
playerID1
```

```python
#this is our base to merge data, the previous had 57*2 rows
y = dfgameLog.merge(dfeventSequence, how="cross").merge(playerID0, how="cross").merge(playerID1, how="cross")
y
```

```python
#rename columns from the merge
z=y.rename(columns={"userID_x":"playerID_0_userID","nickname_x":"playerID_0_nickname",
                  "userID_y":"playerID_1_userID","nickname_y":"playerID_1_nickname"})
z
```

```python
#rearrange the columns in df z2
z2=z[['sessionID','teamID','playerID_0_userID','playerID_0_nickname','playerID_1_userID','playerID_1_nickname','attemptID',
      'sessionExtData',
      'eventName','eventStartTime','eventEndTime','eventBy','eventTo','eventResult','eventLocation','eventExtData']]
z2
```

```python
# List1 - extracting the sesssionExtData 
lst = [[2, 'random', 'jiangang','hao'],[2, 'random', 'jiangang','hao'],
[2, 'random', 'jiangang','hao'],[2, 'random', 'jiangang','hao']]

    
df = pd.DataFrame(lst, columns =['n_player', 'team_assembly_mode', 'P1','P2'])
extra=df.head(1)
extra
```

```python
#merge these variables into zfinal
zfinal=z2.merge(extra, how="cross")
zfinal
```

```python
#rearrange the file, delete extra columns not needed
zfinal=zfinal[['sessionID','teamID','playerID_0_userID','playerID_0_nickname','playerID_1_userID','playerID_1_nickname','attemptID',
      'n_player','team_assembly_mode','P1','P2',
      'eventName','eventStartTime','eventEndTime','eventBy','eventTo','eventResult','eventLocation','eventExtData']]
zfinal
```

```python
pd.DataFrame.to_csv(zfinal, "week3_homework_JSON.csv", sep=',')
```
