```python
#Load in libraries that will be needed
import pandas as pd
import nltk
import numpy as np
```

```python
#Read in dataset into a dataframe, it is tab-delimited
df = pd.read_csv("PSTRE_syntheticData.csv", sep='\t',)
df
```

```python
#PART 1 - Create two new variables: action sequence variable and time interval sequence for each participant
#sort the data by TestTakerID and then by TimeStamp.  It may already be, but just in case
#we need it sorted to determing the Action Sequence
df.sort_values(by=['TestTakerID','Timestamp'])
df
```

```python
#Create Action Sequence variable
#Use groupby function, based on the TestTakerID (unique ID), and cumulatively count
df["ActionSequence"]=df.groupby(['TestTakerID']).cumcount()
df
```

```python
#Create a Time Elapsed variable, then merge it into the file using the "join" function
TimeElapse=df['Timestamp'].diff(periods=1).to_frame().rename(columns={"Timestamp":"TimeElapse"})
dfnew=df.join(TimeElapse)
dfnew
```

```python
#We now have some negative times in there, which occur when we sum the prior
#higher time with the first action time.  This function replaces
#a negative value of elapsed time with NaN
dfnew.loc[(dfnew['TimeElapse'] <= 0), 'TimeElapse']=np.nan
#take a look at some records.  Now we have the Action Sequence and the Time Elapsed
dfnew.head(40)
```

```python
#PART TWO - Extract the time for the first action for each participant
#Create a dataset with just the first actions
#drop the Timestamp column, it is no longer applicable
firstactions=dfnew[dfnew["ActionSequence"] == 1]
firstactions=firstactions.reset_index(drop=True).rename(columns={"TimeElapse":"TimeToFirstAction"}).drop(columns={"Timestamp"})
#N=1079 student first actions
firstactions
```

```python
#PART THREE Represent each action sequence by n-grams (n=2)
#create a dataset called ngrams using the original "df" dataset with all actions
#use the groupby function based on the "coding value", which is the action, they are merged with spaces
#separating them.  Index is reset too
ngrams_df = df.groupby(['TestTakerID'])['Coding'].apply(' '.join).reset_index()
ngrams_df
```

```python
#IMPORT NLTK to produce tokens.  I may not need the funtion or "punkt"
import nltk
from nltk.util import ngrams
nltk.download('punkt')
 
# Function to generate n-grams from sentences.
def extract_ngrams(data, num):
    n_grams = ngrams(nltk.word_tokenize(data), num)
    return [ ' '.join(grams) for grams in n_grams]
 
data = 'A class is a blueprint for the object.'
 
print("1-gram: ", extract_ngrams(data, 1))
print("2-gram: ", extract_ngrams(data, 2))
print("3-gram: ", extract_ngrams(data, 3))
print("4-gram: ", extract_ngrams(data, 4))
```

```python
#Extract bigrams from Path and generate Bigrams
ngram2 = ngrams_df['Coding'].apply(lambda row: list(nltk.bigrams(row.split(' ')))).to_frame().rename(columns={"Coding": "Ngrams2"})
ngram2
```

```python
# Merge back into the dataset, then print the first 40 records to check them out.  These are the n-grams (n=2)
ngrams_df = ngrams_df.join(ngram2)
ngrams_df.head(40)
```
